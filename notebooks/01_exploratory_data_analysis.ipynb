# FPL Data Exploratory Data Analysis

This notebook explores the engineered FPL dataset to verify features and understand distributions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting style
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
# Load processed features
df = pd.read_csv('../data/processed/engineered_features.csv')
df.head()
```

## Verifying Lag Features
Let's look at a specific player to ensure our lag and rolling features are calculated correctly chronologically.

```python
kane = df[df['name'] == 'Harry Kane'].sort_values('kickoff_time')
cols_to_show = ['GW', 'kickoff_time', 'total_points', 'target_next_gw_points', 'lag_1_total_points', 'rolling_3_avg_total_points']
kane[cols_to_show].head(10)
```

## Target Variable Distribution
FPL points are heavily skewed towards 0, 1, and 2.

```python
sns.countplot(x='target_next_gw_points', data=df[df['target_next_gw_points'] <= 15])
plt.title('Distribution of FPL Points (Next GW)')
plt.xlabel('Points')
plt.ylabel('Count')
plt.show()
```
